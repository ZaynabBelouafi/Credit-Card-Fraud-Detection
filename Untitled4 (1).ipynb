```python
import pandas as pd

print(" Lecture intelligente du fichier...")

# 1. On lit en disant qu'il y a un header (ligne 0)
# quotechar='"' dit à Pandas d'ignorer les guillemets autour des titres et de la Classe
df = pd.read_csv("/content/creditcard.csv", header=0, sep=',', quotechar='"')

# 2. Petit nettoyage de sécurité
# Parfois, les guillemets restent dans les noms de colonnes, on les vire
df.columns = df.columns.str.replace('"', '').str.strip()

# 3. La colonne 'Class' est parfois lue comme du texte à cause des guillemets ("0")
# On force la conversion en nombres
df['Class'] = pd.to_numeric(df['Class'], errors='coerce')

# 4. Vérification finale
print(f"\n TERMINÉ !")
print(f"Lignes   : {df.shape[0]}")
print(f"Colonnes : {df.shape[1]} (Doit être 31)")
print("-" * 30)
print(df.head())
df.info()
```

```python
df = df.dropna()
df["Class"] = df["Class"].astype(int)
```

```python
df["Class"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Class", data=df)
plt.title("Distribution des transactions frauduleuses")
plt.show()
```

```python
sns.histplot(df["Amount"], bins=50)
plt.title("Distribution des montants")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Séparer X et y
X = df.drop("Class", axis=1)
y = df["Class"]

# Split TRAIN / TEST
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Standardisation (IMPORTANT : après split)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(" Split and scaling done")
print(X_train.shape, X_test.shape)

```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42, k_neighbors=3)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print(y_train_res.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000, n_jobs=-1)
lr.fit(X_train_res, y_train_res)

y_pred_lr = lr.predict(X_test)
y_prob_lr = lr.predict_proba(X_test)[:, 1]
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train_res, y_train_res)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:, 1]
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train_res, y_train_res)

y_pred_rf = rf.predict(X_test)
y_prob_rf = rf.predict_proba(X_test)[:, 1]
```

```python
!pip install xgboost
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    random_state=42,
    eval_metric="logloss"
)

xgb.fit(X_train_res, y_train_res)

y_pred_xgb = xgb.predict(X_test)
y_prob_xgb = xgb.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Random Forest")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

fpr, tpr, _ = roc_curve(y_test, y_prob_rf)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f"Random Forest (AUC = {roc_auc:.4f})")
plt.plot([0,1],[0,1],'--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def eval_model(name, y_true, y_pred):
    return {
        "Model": name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred),
        "Recall": recall_score(y_true, y_pred),
        "F1": f1_score(y_true, y_pred)
    }

results = []
results.append(eval_model("Logistic Regression", y_test, y_pred_lr))
results.append(eval_model("Random Forest", y_test, y_pred_rf))
results.append(eval_model("XGBoost", y_test, y_pred_xgb))

import pandas as pd
pd.DataFrame(results)
```
